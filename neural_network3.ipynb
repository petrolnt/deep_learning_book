```python
#three inputs, three weights, one output
import matplotlib.pyplot as plt

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]
win_or_lose_binary = [1, 1, 0, 1]

true = win_or_lose_binary[0]
pred = 0
alpha = 0.01

weights = [0.1, 0.2, -0.1]

input = [toes[0], wlrec[0], nfans[0]]

def neural_network(input, weights):
    out = 0
    assert(len(input) == len(weights))
    for i in range(len(input)):
        out += input[i] * weights[i]
    return out

def ele_mul(scalar, vector):
    out = []
    for i in range(len(vector)):
        out.append(vector[i] * scalar)
    return out
    

iter = 0
graph_errors = []
graph_weights0 = []
graph_weights1 = []
graph_weights2 = []
for iter in range(300):
    pred = neural_network(input, weights)
    error = (pred - true) ** 2
    delta = pred - true
    iter += 1
    
    
    
    weight_deltas = ele_mul(delta, input)
    weight_deltas[0] = 0
    print("Iteration: " + str(iter))
    print("Pred: " + str(pred))
    print("Error: " + str(error))
    print("Delta: " + str(delta))
    print("Weights: " + str(weights))
    print("Weight_Deltas: " + str(weight_deltas))
    print
    
    for i in range(len(weights)):
        weights[i] -= weight_deltas[i] * alpha
        
    graph_weights0.append(weights[0])
    graph_weights1.append(weights[1])
    graph_weights2.append(weights[2])
    graph_errors.append(error)
    
plt.figure(figsize=(15,5))
#plt.axis([0.08, 0.205, 0, 0.025])
plt.ylabel('Error')
plt.xlabel('Weight')
plt.plot(graph_weights0, graph_errors,'ro', graph_weights0, graph_errors,'r-')
plt.plot(graph_weights1, graph_errors,'go', graph_weights1, graph_errors,'g-')
plt.plot(graph_weights2, graph_errors,'bo', graph_weights2, graph_errors,'b-')

plt.show()
        
```
